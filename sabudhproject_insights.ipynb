```python
import pandas as pd
```

```python
books = pd.read_csv("/content/books.csv", encoding='windows-1250', delimiter=';', quotechar="'", error_bad_lines=False)
ratings = pd.read_csv("/content/ratings.csv", encoding='windows-1250', delimiter=';', quotechar="'", error_bad_lines=False)
user = pd.read_csv("/content/users.csv", encoding='windows-1250', delimiter=';', quotechar="'", error_bad_lines=False)
```

## Data Preprocessing

```python
# Remove double quotes from column names
books.columns = [col.strip('"') for col in books.columns]
user.columns = [col.strip('"') for col in user.columns]
ratings.columns = [col.strip('"') for col in ratings.columns]
```

```python
# List of DataFrames
dataframes = [user, books, ratings]

# Iterate through each DataFrame and each column
for df in dataframes:
    for col in df.columns:
        if df[col].dtype == object:
            df[col] = df[col].str.replace('"', '')
```

## Find the 10 trending books

```python
# Calculate the average rating for each book
average_ratings = ratings.groupby('ISBN')['Book-Rating'].mean()

# Get the top 10 books with the highest average ratings
top_trending_books_isbn = average_ratings.nlargest(10).index

# Get the corresponding book titles from the 'books' DataFrame
trending_books_titles = books[books['ISBN'].isin(top_trending_books_isbn)]['Book-Title']

# Display the top 10 trending book titles
print(trending_books_titles)
```

## Find the most active_user

```python
most_active_user = ratings['User-ID'].value_counts().idxmax()
most_active_user
```

## Location of the most active user

```python
most_active_geo_area = user['Location'].value_counts().idxmax()
most_active_geo_area
```

## Find trending books in a particular geographical area

```python
# Specify the location for which you want to find trending books
geo_area = "london"

# Remove rows with NaN values in the 'Location' column of the 'user' DataFrame
user = user.dropna(subset=['Location'])

# Filter the 'ratings' DataFrame to include only ratings from users in the specified location
trending_books_in_geo = ratings[ratings['User-ID'].isin(user[user['Location'].str.contains(geo_area)]['User-ID'])]

# Calculate the average rating for each book
average_ratings = trending_books_in_geo.groupby('ISBN')['Book-Rating'].mean()

# Get the top 10 books with the highest average ratings
top_trending_books_isbn = average_ratings.nlargest(10).index

# Get the book titles of the top trending books
trending_books = books[books['ISBN'].isin(top_trending_books_isbn)]['Book-Title']
trending_books

#we prefer average to avoid the manipulation of the highest values - bias!!
```
